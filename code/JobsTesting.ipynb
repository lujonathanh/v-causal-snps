```python
import numpy as np
import pickle
import csv

filenames = []
for i in range(10):
    matr1 = np.zeros((10,10))
    matr1[i] = np.ones(10)
    
    matr2 = np.zeros((10,10))
    matr2[i] = -1 * np.ones(10)
    
    matr3 = 3 * matr2
    
    filename = "tuple"  + str(i) + ".p"
    pickle.dump((matr1, matr2, matr3), open(filename, 'w'))
    print "Written to ", filename
    filenames.append(filename)

with open("output_list.txt", 'w') as csvfile:
    writer = csv.writer(csvfile, delimiter='\t')
    for filename in filenames:
        writer.writerow([filename])
```
